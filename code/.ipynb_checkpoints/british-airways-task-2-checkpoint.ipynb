## <span id=2> 2. Data Preporcessing & Feature Engineering</span>  

```python
day_to_num = {'Sat':6, 'Wed':3, 'Thu':2, 'Mon':1, 'Sun':0, 'Tue':4, 'Fri':5}
df['flight_day'] = df['flight_day'].map(day_to_num)

df['route_from'] = np.vectorize(lambda x: x[:3], otypes=[str])(df['route'])
df['route_to'] = np.vectorize(lambda x: x[3:], otypes=[str])(df['route'])
```

```python
temp_cols = ['booking_origin', 'num_passengers', 'purchase_lead', 'flight_hour', 'flight_duration']
num_cols = ['num_passengers', 'purchase_lead', 'flight_hour', 'flight_duration']

for num_col in num_cols:
    temp_group = df[['booking_origin', num_col]].groupby('booking_origin')[num_col].agg(['mean', 'sum', 'count'])

    temp_group.columns = [col + '_' + num_col +'_origin' for col in temp_group.columns]
    df = pd.merge(df, temp_group, how='inner', left_on='booking_origin', right_index=True)
```

```python
temp_describe = df[['num_passengers', 'purchase_lead', 'length_of_stay', 'booking_complete']].describe([.999])
```

```python
# remove outliers - 99.9 percentile trim

temp_cols = ['num_passengers', 'purchase_lead', 'length_of_stay', 'booking_complete']

for col in temp_cols:
    drop_index = df[[col]][df[col] > temp_describe[col]['99.9%']].index
    df.drop(labels=drop_index, axis=0, inplace=True)
    df.reset_index(drop=True, inplace=True)
```

## <span id=3> 3. Feature Selection</span>  

```python
df.columns
```

```python
# remove unnecessary columns
temp_cols = ['route', 'trip_type']
df.drop(labels=temp_cols, axis=1, inplace=True)
```

```python
df_cat = df.select_dtypes(exclude='number')
df_num = df.select_dtypes(include='number')

df_cat_dummy = pd.get_dummies(df_cat, drop_first=True)
df_dummy = pd.concat([df_num, df_cat_dummy], axis=1)
```

## <span id=4> 4. Models Training</span>  

```python
from sklearn.model_selection import train_test_split, RandomizedSearchCV, StratifiedKFold
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from imblearn.over_sampling import SMOTE
from imblearn.pipeline import Pipeline as imb_Pipeline

from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier

from sklearn.metrics import precision_score, accuracy_score, recall_score, \
f1_score, roc_auc_score, accuracy_score, classification_report, plot_confusion_matrix
```

```python
X = df_dummy.drop('booking_complete', axis=1)
y = df_dummy['booking_complete']

scaler = StandardScaler()
```

### Logistic Regression Model

```python
accuracy_lst = []
precision_lst = []
recall_lst = []
f1_lst = []
auc_lst = []

smote = SMOTE(sampling_strategy='minority', random_state=1, n_jobs=-1)

lr = LogisticRegression(solver='sag', n_jobs=-1, verbose=10, random_state=1)
lr_params = {'lr__C': [0.01, 1, 100]}

sss = StratifiedKFold(n_splits=5, random_state=1, shuffle=True)

pipe = [('smote', smote), ('lr', lr)]
```

```python
pipeline = imb_Pipeline(pipe)

rand_lr = RandomizedSearchCV(pipeline, lr_params, 
                             n_iter=4, 
                             n_jobs=-1, 
                             verbose=10, 
                             scoring='recall', 
                             random_state=1)

for train, test in sss.split(X, y):
    
    scaled_X_train = scaler.fit_transform(X.iloc[train])
    scaled_X_test = scaler.transform(X.iloc[test])
    
    rand_lr.fit(scaled_X_train, y.iloc[train])
    y_hat = rand_lr.predict(scaled_X_test)
    
    accuracy_lst.append(accuracy_score(y.iloc[test], y_hat))    
    precision_lst.append(precision_score(y.iloc[test], y_hat))    
    recall_lst.append(recall_score(y.iloc[test], y_hat))
    f1_lst.append(f1_score(y.iloc[test], y_hat))
    auc_lst.append(roc_auc_score(y.iloc[test], y_hat))
```

```python
print(f'accuracy CV: {np.array(accuracy_lst).mean().round(4)}')
print(f'precision CV: {np.array(precision_lst).mean().round(4)}')
print(f'recall CV: {np.array(recall_lst).mean().round(4)}')
print(f'f1 CV: {np.array(f1_lst).mean().round(4)}')
print(f'acu CV: {np.array(auc_lst).mean().round(4)}')
```

```python
rand_lr.best_estimator_
```

```python
# second round to search hyperparameter

accuracy_lst = []
precision_lst = []
recall_lst = []
f1_lst = []
auc_lst = []

smote = SMOTE(sampling_strategy='minority', random_state=1, n_jobs=-1)

lr = LogisticRegression(solver='sag', n_jobs=-1, verbose=10, random_state=1)
lr_params = {'lr__C': [100, 500, 1000]}

sss = StratifiedKFold(n_splits=5, random_state=1, shuffle=True)

pipe = [('smote', smote), ('lr', lr)]
```

```python
pipeline = imb_Pipeline(pipe)

rand_lr = RandomizedSearchCV(pipeline, lr_params, 
                             n_iter=4, 
                             n_jobs=-1, 
                             verbose=10, 
                             scoring='recall', 
                             random_state=1)

for train, test in sss.split(X, y):
    
    scaled_X_train = scaler.fit_transform(X.iloc[train])
    scaled_X_test = scaler.transform(X.iloc[test])
    
    rand_lr.fit(scaled_X_train, y.iloc[train])
    y_hat = rand_lr.predict(scaled_X_test)
    
    accuracy_lst.append(accuracy_score(y.iloc[test], y_hat))    
    precision_lst.append(precision_score(y.iloc[test], y_hat))    
    recall_lst.append(recall_score(y.iloc[test], y_hat))
    f1_lst.append(f1_score(y.iloc[test], y_hat))
    auc_lst.append(roc_auc_score(y.iloc[test], y_hat))
```

```python
print(f'accuracy CV: {np.array(accuracy_lst).mean().round(4)}')
print(f'precision CV: {np.array(precision_lst).mean().round(4)}')
print(f'recall CV: {np.array(recall_lst).mean().round(4)}')
print(f'f1 CV: {np.array(f1_lst).mean().round(4)}')
print(f'acu CV: {np.array(auc_lst).mean().round(4)}')
```

```python
rand_lr.best_estimator_
```

```python
lr_final = LogisticRegression(C=100,
                        solver='sag', 
                        n_jobs=-1, 
                        verbose=10, 
                        random_state=1)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1, stratify=y)

scaled_X_train = scaler.fit_transform(X_train)
scaled_X_test = scaler.transform(X_test)

X_resampled_train, y_resampled_train = smote.fit_resample(scaled_X_train, y_train)

lr_final.fit(X_resampled_train, y_resampled_train)
```

```python
print(classification_report(y_test, lr_final.predict(scaled_X_test)))
```

```python
roc_auc_score(y_test, lr_final.predict(scaled_X_test))
```

### Random Forest Model

```python
accuracy_lst = []
precision_lst = []
recall_lst = []
f1_lst = []
auc_lst = []

smote = SMOTE(sampling_strategy='minority', random_state=1, n_jobs=-1)

rf = RandomForestClassifier(n_jobs=-1, verbose=10, random_state=1)
rf_params = {'rf__max_depth': [4,6,8]}

sss = StratifiedKFold(n_splits=5, random_state=1, shuffle=True)

pipe = [('smote', smote), ('rf', rf)]
```

```python
pipeline = imb_Pipeline(pipe)

rand_rf = RandomizedSearchCV(pipeline, rf_params, 
                             n_iter=4, 
                             n_jobs=-1, 
                             verbose=10, 
                             scoring='recall', 
                             random_state=1)

for train, test in sss.split(X, y):
    
    scaled_X_train = scaler.fit_transform(X.iloc[train])
    scaled_X_test = scaler.transform(X.iloc[test])
    
    rand_rf.fit(scaled_X_train, y.iloc[train])
    y_hat = rand_rf.predict(scaled_X_test)
    
    accuracy_lst.append(accuracy_score(y.iloc[test], y_hat))    
    precision_lst.append(precision_score(y.iloc[test], y_hat))    
    recall_lst.append(recall_score(y.iloc[test], y_hat))
    f1_lst.append(f1_score(y.iloc[test], y_hat))
    auc_lst.append(roc_auc_score(y.iloc[test], y_hat))
```

```python
print(f'accuracy CV: {np.array(accuracy_lst).mean().round(4)}')
print(f'precision CV: {np.array(precision_lst).mean().round(4)}')
print(f'recall CV: {np.array(recall_lst).mean().round(4)}')
print(f'f1 CV: {np.array(f1_lst).mean().round(4)}')
print(f'acu CV: {np.array(auc_lst).mean().round(4)}')
```

```python
rand_rf.best_estimator_
```

```python
rf_final = RandomForestClassifier(max_depth=4, n_jobs=-1, verbose=10, random_state=1)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1, stratify=y)

scaled_X_train = scaler.fit_transform(X_train)
scaled_X_test = scaler.transform(X_test)

X_resampled_train, y_resampled_train = smote.fit_resample(scaled_X_train, y_train)

rf_final.fit(X_resampled_train, y_resampled_train)
```

```python
print(classification_report(y_test, rf_final.predict(scaled_X_test)))
```

```python
roc_auc_score(y_test, rf_final.predict(scaled_X_test))
```
